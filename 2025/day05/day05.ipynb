```python
import itertools
```

```python
def preprocess(filename):
    file = open(filename, 'r')
    data = file.read()

    ingredient_db = data.split('\n\n')
    ingredient_db = [_.split() for _ in ingredient_db]

    ingredient_db[0] = [_.split('-') for _ in ingredient_db[0]]
    ingredient_db[0] = [range(int(_[0]), int(_[1])+1) for _ in ingredient_db[0]]
    ingredient_db[1] = list(map(int, ingredient_db[1]))

    return ingredient_db
print(preprocess('example.txt'))
print(len(preprocess('example.txt')[0][0]))
```

```python
def part_1(ingredient_db):
    certified_fresh = 0
    for ingredient_id in ingredient_db[1]:
        for fresh_range in ingredient_db[0]:
            if ingredient_id in fresh_range:
                certified_fresh += 1
                break
    return certified_fresh

print(part_1(preprocess('example.txt')))
print(part_1(preprocess('aocd5_input.txt')))
```

## pt 2

```python
def part_2(ingredient_db):
    fresh_tups = [(_[0], _[-1]) for _ in ingredient_db[0]]
    fresh_tups.sort()

    fresh_ids = [range(fresh_tups[0][0], fresh_tups[0][1]+1)]
    fresh_tups.pop(0)

    while len(fresh_tups) > 0:
        # test for new range either above or below
        if fresh_tups[0][0] > fresh_ids[-1][-1]: # min is above largest
            fresh_ids.append(range(fresh_tups[0][0], fresh_tups[0][1]+1))
        elif fresh_tups[0][1] < fresh_ids[0][0]: # max is below lowest
            fresh_ids.insert(0, range(fresh_tups[0][0], fresh_tups[0][1]+1))
        else:
            # test if overlap
            start_overlap = -1
            end_overlap = -1
            for fresh_id in range(len(fresh_ids)):
                if fresh_tups[0][0] in fresh_ids[fresh_id]:
                    start_overlap = fresh_id
                if fresh_tups[0][1] in fresh_ids[fresh_id]:
                    end_overlap = fresh_id

            if start_overlap != -1 and end_overlap == -1: # start in but end not
                fresh_ids[start_overlap] = range(fresh_ids[start_overlap][0], fresh_tups[0][1]+1)
            elif end_overlap != -1 and start_overlap == -1: # end in but not start
                fresh_ids[end_overlap] = range(fresh_tups[0][0], fresh_ids[end_overlap][-1]+1)
            elif end_overlap != -1 and start_overlap != -1 and end_overlap != start_overlap: # if the range spans two ranges
                fresh_ids[start_overlap] = range(fresh_ids[start_overlap][0], fresh_ids[end_overlap][-1]+1)
                fresh_ids.pop(end_overlap)
        fresh_tups.pop(0)

    certified_fresh = 0
    for fresh_range in fresh_ids:
        certified_fresh += len(fresh_range)
    return certified_fresh

print(part_2(preprocess('example.txt')))
print(part_2(preprocess('aocd5_input.txt')))
```

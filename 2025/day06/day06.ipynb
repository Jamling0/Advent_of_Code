```python
import math
```

```python
def parse_pt_1(filename):
    file = open(filename, 'r')
    data = file.read()

    math_homework = data.splitlines()
    math_homework = [_.split() for _ in math_homework]

    math_homework[0], math_homework[-1] = math_homework[-1], math_homework[0]

    problem_list = []
    problem_ctr = 0
    while problem_ctr < len(math_homework[0]):
        num_list = []
        for homework in range(1, len(math_homework)):
            num_list.append(int(math_homework[homework][problem_ctr]))
        problem_list.append(num_list)
        problem_ctr += 1

    operation_list = math_homework[0]

    return problem_list, operation_list

print(parse_pt_1('example.txt'))
print(parse_pt_1('aocd6_input.txt'))
```

```python
def calculate(num_list, operation_list):
    sum_answers = 0
    for homework in range(len(num_list)):
        if operation_list[homework] == '+':
            sum_answers += sum(num_list[homework])
        else:
            sum_answers += math.prod(num_list[homework])

    return sum_answers

# print(part_1(preprocess('example.txt')[0], preprocess('example.txt')[1]))
print(calculate(parse_pt_1('aocd6_input.txt')[0], parse_pt_1('aocd6_input.txt')[1]))
```

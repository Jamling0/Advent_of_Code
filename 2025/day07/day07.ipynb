## pt 1

```python
def preprocess(filename):
    file = open(filename, 'r')
    data = file.read()

    manifold_diagram = data.splitlines()
    manifold_diagram = [list(_) for _ in manifold_diagram]

    return manifold_diagram

print(preprocess('example.txt'))
# print(preprocess('aocd7_input.txt'))
```

```python
def part_1(manifold_diagram):
    beam_start_index = manifold_diagram[0].index('S')
    beam_pos = {beam_start_index}
    num_splits = 0
    for row in range(2, len(manifold_diagram), 2):
        for col in range(len(manifold_diagram[row])):
            if manifold_diagram[row][col] == '^' and col in beam_pos:
                beam_pos.remove(col)
                beam_pos.add(col - 1)
                beam_pos.add(col + 1)
                num_splits += 1
    return num_splits

print(part_1(preprocess('example.txt')))
print(part_1(preprocess('aocd7_input.txt')))
```

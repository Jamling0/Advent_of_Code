## pt 1

```python
def preprocess(filename):
    file = open(filename, 'r')
    data = file.read()

    warehouse_floor = data.split()
    warehouse_floor = [list(_) for _ in warehouse_floor]

    return warehouse_floor
print(preprocess('example.txt'))
```

```python
def part_1(warehouse_floor):
    pickable_rolls = 0
    for row in range(len(warehouse_floor)):
        for column in range(len(warehouse_floor[row])):
            if warehouse_floor[row][column] == '@':
                adj_pos_occupied = 0
                if row == 0:
                    row_ptrs = [0, 1]
                elif row == len(warehouse_floor) - 1:
                    row_ptrs = [-1, 0]
                else:
                    row_ptrs = [-1, 0, 1]
                if column == 0:
                    column_ptrs = [0, 1]
                elif column == len(warehouse_floor[row]) - 1:
                    column_ptrs = [-1, 0]
                else:
                    column_ptrs = [-1, 0, 1]
                for row_check in row_ptrs:
                    for column_check in column_ptrs:
                        if warehouse_floor[row + row_check][column + column_check] == '@':
                            adj_pos_occupied += 1
                if adj_pos_occupied - 1 < 4:
                    pickable_rolls += 1

                    # if warehouse_floor[row-1][column-1] == '@': # up-left corner
                    #     adj_pos_occupied += 1
                    # if warehouse_floor[row-1][column] == '@': # up
                    #     adj_pos_occupied += 1
                    # if warehouse_floor[row-1][column+1] == '@': # up-right corner
                    #     adj_pos_occupied += 1
                    # if warehouse_floor[row][column-1] == '@': # left
                    #     adj_pos_occupied += 1


    return pickable_rolls
print(part_1(preprocess('example.txt')))
print(part_1(preprocess('aocd4_input.txt')))
```

## pt 2

```python
def part_2(warehouse_floor):
    removeable_rolls = []
    removeable_rolls_ctr = 0
    rolls_to_be_removed = True
    while rolls_to_be_removed:
        for row in range(len(warehouse_floor)):
            for column in range(len(warehouse_floor[row])):
                if warehouse_floor[row][column] == '@':
                    adj_pos_occupied = 0
                    if row == 0:
                        row_ptrs = [0, 1]
                    elif row == len(warehouse_floor) - 1:
                        row_ptrs = [-1, 0]
                    else:
                        row_ptrs = [-1, 0, 1]
                    if column == 0:
                        column_ptrs = [0, 1]
                    elif column == len(warehouse_floor[row]) - 1:
                        column_ptrs = [-1, 0]
                    else:
                        column_ptrs = [-1, 0, 1]
                    for row_check in row_ptrs:
                        for column_check in column_ptrs:
                            if warehouse_floor[row + row_check][column + column_check] == '@':
                                adj_pos_occupied += 1
                    if adj_pos_occupied - 1 < 4:
                        removeable_rolls.append((row, column))
                        # removeable_rolls_ctr += 1
        if len(removeable_rolls) == 0:
            rolls_to_be_removed = False

        while len(removeable_rolls) > 0:
            warehouse_floor[removeable_rolls[0][0]][removeable_rolls[0][1]] = 'x'
            removeable_rolls.remove((removeable_rolls[0][0], removeable_rolls[0][1]))
            removeable_rolls_ctr += 1

    return removeable_rolls_ctr

print(part_2(preprocess('example.txt')))
print(part_2(preprocess('aocd4_input.txt')))
```
